## Q4. What is ‘DataFrame’ in pandas and how is it different from pandas.series? Explain with an example.
### Answer:-

### In PANDAS a DataFrame is a 2-dimensional labeled data structure with columns of potentially different types. It is a tabular data structure that consists of rows and columns, where each column can have a different data type (e.g. string, integer, float) and can have a name or label.

### On the other hand, a Pandas Series is a one-dimensional labeled array that can hold any data type (e.g. string, integer, float, object). A Series is a single column of data with an associated label, called the index.

## Q5. What are some common functions you can use to manipulate data in a Pandas DataFrame? Can you give an example of when you might use one of these functions?
### Answer:- 

### Pandas provides a wide range of functions to manipulate data in a DataFrame. Some common functions include:
### 1. - head() and tail(): These functions can be used to display the first or last few rows of a DataFrame, respectively. 
### 2. - describe(): This function generates a statistical summary of the DataFrame, including count, mean, standard deviation, and quartile values.
### 3. - groupby(): This function can be used to group the data in the DataFrame based on a specific column or multiple columns, and then perform aggregate functions on the resulting groups. 
### 4. - sort_values(): This function can be used to sort the data in the DataFrame based on one or more columns.
### 5. - dropna(): This function can be used to remove any rows from the DataFrame that contain missing or null values.

## Q6. Which of the following is mutable in nature Series, DataFrame, Panel?
### Answer:-

### In Pandas, both Series and DataFrame are mutable in nature, while Panel is considered deprecated and is no longer recommended for use.

### A mutable object is an object whose state can be changed after it is created. This means that we can modify the data in a Series or DataFrame after it is created by assigning new values to individual cells or adding or removing rows and columns.
